```python
import speech_recognition as sr
import pyttsx3
import datetime
import webbrowser
import openai
import subprocess
import os
import qrcode
import gradio as gr
import numpy as np
import noisereduce as nr

# Update your API key here
openai.api_key = "sk-proj-5KlViEpxSRaCoCA29deYT3BlbkFJDxy5CFuklZt2GTNFhhwP"

def speak(text):
    engine = pyttsx3.init()
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[1].id)  # change to voice index 1 for female voice
    engine.say(text)
    engine.runAndWait()

def listen():
    recognizer = sr.Recognizer()
    microphone = sr.Microphone()
    
    with microphone as source:
        recognizer.adjust_for_ambient_noise(source)
        print("Say something:")
        audio = recognizer.listen(source)
        
    # Convert the audio to numpy array
    audio_data = np.frombuffer(audio.get_raw_data(), np.int16)
    
    # Perform noise reduction
    reduced_noise = nr.reduce_noise(y=audio_data, sr=16000)
    
    # Convert the reduced noise numpy array back to audio
    audio_reduced = sr.AudioData(reduced_noise.tobytes(), audio.sample_rate, audio.sample_width)
    
    try:
        text = recognizer.recognize_google(audio_reduced, language='en-US')
        print("You said: {}".format(text))
    except sr.UnknownValueError:
        print("Sorry, I could not understand what you said")
        text = None
    except sr.RequestError as e:
        print("Error accessing the Google Speech Recognition service: {0}".format(e))
        text = None
        
    return text

def perform_task(text):
    text = text.lower()
    if "hey friday" in text:
        speak("Hello! How can I help you today?")
        return "Hello! How can I help you today?"
    elif "what time is it" in text:
        now = datetime.datetime.now()
        response = "The current time is {}:{}:{}".format(now.hour, now.minute, now.second)
        speak(response)
        return response
    elif "search" in text:
        search_term = text.replace("search", "").strip()
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": "search " + search_term}]
        ).choices[0].message['content']
        print(response)
        speak("According to my search, " + response)
        return response
    elif "play video on youtube" in text:
        search_term = text.replace("play video on youtube", "").strip()
        url = "https://www.youtube.com/results?search_query=" + search_term
        webbrowser.open(url)
        response = "Playing the video on youtube for {}".format(search_term)
        speak(response)
        return response
    elif "play song on spotify" in text:
        search_term = text.replace("play song on spotify", "").strip()
        url = "https://open.spotify.com/search/" + search_term
        webbrowser.open(url)
        response = "Playing the song on spotify for {}".format(search_term)
        speak(response)
        return response
    elif "find" in text:
        search_term = text.replace("find", "").strip()
        url = "https://google.com/search?q=" + search_term
        webbrowser.open(url)
        response = "Here's your search result for {}".format(search_term)
        speak(response)
        return response
    elif "stop" in text:
        response = "Time period of me is over, thank you sir."
        speak(response)
        return response
    elif "open notepad" in text:
        subprocess.call(["notepad.exe"])
        response = "Opening Notepad."
        speak(response)
        return response
    elif "open spotify" in text:
        subprocess.call(["spotify.exe"])
        response = "Opening Spotify."
        speak(response)
        return response
    elif "open asphalt" in text:
        subprocess.call(["asphalt 9 legends.exe"])
        response = "Opening Asphalt 9 Legends."
        speak(response)
        return response
    elif "open calculator" in text:
        subprocess.call(["calc.exe"])
        response = "Opening Calculator."
        speak(response)
        return response
    elif "open paint" in text:
        subprocess.call(["mspaint.exe"])
        response = "Opening Paint."
        speak(response)
        return response
    elif "shutdown" in text:
        os.system("shutdown /s /t 1")
        response = "Shutting down the PC."
        speak(response)
        return response
    elif "restart" in text:
        subprocess.call(["shutdown", "/r", "/t", "1"])
        response = "Restarting your device. I will be back once the device gets booted."
        speak(response)
        return response
    elif "lock" in text:
        os.system("rundll32.exe user32.dll,LockWorkStation")
        response = "Locking your device."
        speak(response)
        return response
    elif "sleep" in text:
        subprocess.call(["rundll32.exe", "powrprof.dll,SetSuspendState"])
        response = "Putting the PC to sleep."
        speak(response)
        return response
    elif "location" in text:
        locate = text.replace("location", "").strip()
        response = f"Searching for {locate} on Google Maps"
        speak(response)
        os.system(f"start chrome https://www.google.com/maps/place/{locate}")
        return response
    elif "qr code" in text:
        qr_content = text.replace("qr code", "").strip()
        img = qrcode.make(qr_content)
        img.save("C://Users//Asus//OneDrive//Pictures//qr code//qrcode.png")
        response = "QR code generated and saved as qrcode.png"
        speak(response)
        return response
    elif "weather " in text:
        city = text.split("in")[-1].strip()
        url = f"https://www.google.com/search?q={city} weather"
        webbrowser.open(url)
        response = "Opening weather in browser."
        speak(response)
        return response
    else:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": text}]
        ).choices[0].message['content']
        speak(response)
        return response

def gradio_interface(text):
    return perform_task(text)

def gradio_voice_interface():
    text = listen()
    if text:
        return text, perform_task(text)
    return "", "Sorry, I could not understand what you said."

with gr.Blocks() as demo:
    gr.Markdown("## JARVIS - Your Personal Voice Assistant")
    
    with gr.Row():
        text_input = gr.Textbox(label="Enter your request")
        text_output = gr.Textbox(label="Response")
    
    text_button = gr.Button("Submit Text Request")
    text_button.click(gradio_interface, inputs=text_input, outputs=text_output)
    
    with gr.Row():
        voice_button = gr.Button("Submit Voice Request")
        voice_input = gr.Textbox(label="Your Voice Input")
        voice_output = gr.Textbox(label="Response")
    
    voice_button.click(gradio_voice_interface, inputs=[], outputs=[voice_input, voice_output])

demo.launch(share=True)
```
